import library yang akan digunakan (tweepy dan csv)

```python
import tweepy
import csv
```

input kode consumer dan secret key yang didapat dari twitter dev app

```python
consumer_key = 'kode anda'
consumer_secret = 'kode anda'
access_token = 'kode anda'
access_token_secret = 'kode anda'
```

autentikasi api

```python
auth = tweepy.auth.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

membuat file csv untuk nanti diisi oleh scrapped tweets

```python
# membuat csv
csvFile = open('queer.csv', 'a')

#mengunakan csv writer
csvWriter = csv.writer(csvFile)

# q = topik yang akan discrap dari search, count = jumlah tweets
for tweet in tweepy.Cursor(api.search, q = "queer", count = 300).items():

    # membuat baris pada csv dengan kolom waktu dan tweetnya
    csvWriter.writerow([tweet.created_at, tweet.text.encode('utf-8')])
    print(tweet.created_at, tweet.text)
csvFile.close()
```
