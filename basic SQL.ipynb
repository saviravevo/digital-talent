```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# Prepare a list of offices sorted by country, state, city
run("SELECT country, state, city FROM offices")
```

```python
# How many employees are there in the company
run("""
    SELECT COUNT(*) AS totalEmployees FROM employees
    """)
```

```python
# What is the total of payments received?
run("""
    SELECT SUM(amount) AS totalPayment FROM payments
    """)
```

```python
# List the product lines that contain 'Cars'
run("""
    SELECT 
    productLine 
    FROM 
    products 
    WHERE 
    productLine LIKE '%Cars'
    """)
```

```python
# Report total payments for October 28, 2004
run("""
    SELECT SUM(amount) AS payment28Oct FROM payments WHERE paymentDate = '2004-10-28'
    """)
```

```python
# Report those payments greater than $100,000
run("""
    SELECT customerNumber,amount FROM payments WHERE amount>100000
    """)
```

```python
# List the products in each product line.
run("SELECT productLine, productName FROM products ORDER BY productLine")
```

```python
# How many products in each product line?
run("""
    SELECT productLine AS 'Product', COUNT(productLine) AS "COUNT" FROM products GROUP BY productLine
    """)
```

```python
# minuimum payment
run("""
SELECT min(amount) 
AS 'Minimum Payment'
from payments
""")
```

```python
# List all payments greater than twice the average payment.
run("""
SELECT *
from payments
where amount >(SELECT AVG(amount) * 2 from payments)
""")
```

```python
# How many distinct products does ClassicModels sell?
run("""
    SELECT productLine, COUNT(productLine) from products GROUP BY productLine
    """)
```

```python
# What is the average percentage markup of the MSRP on buyPrice?
run("SELECT AVG(MSRP - buyPrice) AS 'Average Mark Up' FROM products")
```

```python
# How many distinct products does ClassicModels sell?
run("SELECT DISTINCT productName FROM products")
```

```python
# Report the name and city of customers who don't have sales representatives?
run("""
    SELECT customerName, city, salesRepEmployeeNumber
    FROM customers
    WHERE salesRepEmployeeNumber IS NULL
    """)
```

```python
# What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

run("""
SELECT CONCAT (firstName, " ", lastName) AS fullName, JobTitle
FROM employees
WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'
""")
```

```python
# Which orders have a value greater than $5,000?
run("""
SELECT orderNumber, SUM(priceEach * quantityOrdered) AS Value
FROM orderdetails
GROUP BY orderNumber
Having Value > 5000
""")
```
