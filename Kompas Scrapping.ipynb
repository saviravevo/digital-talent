import lib yang dibutuhkan

```python
import requests
from bs4 import BeautifulSoup
```

memasukkan link yang akan di scrap, masukkan pada variabel url, variabel r digunakan untuk mendapatkan/request web page yang diinput pada variabel url

```python
url = "https://nasional.kompas.com/read/2018/10/28/22394061/ridwan-kamil-bertekad-tak-ada-lagi-kepala-daerah-di-jabar-yang-korup"
r = requests.get(url)
```

menggunakan beuaitufl soup untuk parsing html page

```python
soup = BeautifulSoup(r.content, 'html.parser') 
print(soup.prettify())
```
